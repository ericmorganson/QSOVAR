# Histograms of Magnitudes vs. Normal Distribution

This notebook aims to see how we must change the error of a histogram of g magnitudes in order to make it a normal distribution.

The plotted magnitudes will be $m_i - \mu$ so that the distribution should be centered around 0, by definition.  As mentioned by Eric Morganson, the error of these should be:
$\sigma_i^2 -  \sigma_{\mu}^2 $

So, essentially, we are trying to find how much $$\frac{m_i - \mu}{\sqrt{\sigma_i^2 - \sigma_{\mu}^2}}$$ differs from a normal distribution.

Also, we are trying to find an "a" and "b" such that the iteratively sigma-clipped distributions would have a standard deviation of 1 and a mean as close to 0 as possible AND such that "b" will not change greatly for each G magnitude bins.

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import astropy.io.fits as pyfit
from scipy import stats  
np.set_printoptions(precision=5)
```

```python
#initialize variables
fits = pyfit.open('/home/sam/Documents/Morganson_research/C1_lc.fits')[1].data

print(fits[3]["LC_FLUX_PSF_G"])
print(len(fits))
print(fits[3]['MEAN_PSF_G'])
```
